## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Step 1- Identifying missing values using isna() method
air_bnb_df.isna().sum()
````

````python
# Step 2- Clean the missing values in each column
# cleaning can be done in two ways either we can drop the column or we can fill the missing value with mean or median
````

````python
#filling null values of name column
air_bnb_df['name'].fillna('Unknown',inplace=True)
air_bnb_df['host_name'].fillna('Unknown', inplace=True)
````

````python
#converting string to date time object
air_bnb_df['last_review']=pd.to_datetime(air_bnb_df['last_review'])
# filling null values with mean value
mean_value= air_bnb_df['last_review'].mean()
air_bnb_df.fillna(mean_value,inplace=True)
````

````python
#total number of reviews aross air bnb listings(name, neighbourhood group, price, minimum nights)
areas_review=air_bnb_df.groupby(['name','neighbourhood_group','price','minimum_nights'])['number_of_reviews'].count().reset_index()
sorting_price=areas_review.sort_values('price',ascending=False)
print(sorting_price)
````

````python
# objective - Our objective is to analyze the distribution of Airbnb listings across different neighborhoods within a city or region.
neighborhood_stats = air_bnb_df.groupby(['neighbourhood']).agg({'room_type':'count', 'price': 'mean'}).reset_index()
print( neighborhood_stats)
````

````python
#average price per neighbourhood
average_price_per_neighbourhood=air_bnb_df.groupby(['neighbourhood']).agg({'price':"mean"}).reset_index()
print(average_price_per_neighbourhood)
````

````python
# Sort neighborhoods based on mean price (descending order)
highest_mean_price_neighborhoods = neighborhood_stats.sort_values(by='price', ascending=False).head(5)
# Sort neighborhoods based on mean price (ascending order)
lowest_mean_price_neighborhoods = neighborhood_stats.sort_values(by='price', ascending=True).head(5)

# Sort neighborhoods based on room type count (descending order)
highest_room_type_count_neighborhoods = neighborhood_stats.sort_values(by='room_type', ascending=False).head(5)
# Sort neighborhoods based on room type count (ascending order)
lowest_room_type_count_neighborhoods = neighborhood_stats.sort_values(by='room_type', ascending=True).head(5)

print("Neighborhoods with the Highest Mean Prices:")
print(highest_mean_price_neighborhoods)

print("\nNeighborhoods with the Lowest Mean Prices:")
print(lowest_mean_price_neighborhoods)

print("\nNeighborhoods with the Highest Room Type Counts:")
print(highest_room_type_count_neighborhoods)

print("\nNeighborhoods with the Lowest Room Type Counts:")
print(lowest_room_type_count_neighborhoods)
````

````python
# objective is to  Compare the mean price of listings across different neighborhoods to identify variations in pricing.
output= air_bnb_df.groupby(['neighbourhood']).agg({'price': 'mean'})
print(output)
````

````python
# analyse total number of reviews across neighbouhood and room_type
grouped_df=air_bnb_df.groupby(["neighbourhood","room_type"]).agg({'number_of_reviews':'count'}).reset_index()
#sort by number of reviews in descending order
final_output=grouped_df.sort_values(by='number_of_reviews',ascending=False).head(10)
print(final_output)
````

````python
# Group the data by neighborhood and room type and count occurrences
neighborhood_room_counts = air_bnb_df.groupby(['neighbourhood', 'room_type']).size().reset_index(name='count')

# Find the most opted neighborhood and room type
most_opted_neighborhood = neighborhood_room_counts.loc[neighborhood_room_counts['count'].idxmax()]

print("Most Opted Neighborhood:", most_opted_neighborhood['neighbourhood'])
print("Most Opted Room Type:", most_opted_neighborhood['room_type'])
````

````python
# count of total room types
count=air_bnb_df['room_type'].value_counts()
print(count)
````

````python
# Finding Busiest hosts
busiest_host=air_bnb_df.groupby(['host_name','neighbourhood_group','room_type'])['minimum_nights'].count().reset_index()
valueable_host=busiest_host.sort_values('minimum_nights',ascending=False).head(7)
valueable_host
````

````python
# Average Price acoss different neighbourhood groups
avg_price= air_bnb_df.groupby(['neighbourhood_group']).agg({'price':'mean'})
print(avg_price)
````

````python
#finding unique neighbourhoods
unique_neighborhoods = average_price_per_neighbourhood['neighbourhood'].unique()
print(unique_neighborhoods)
# Step 2: Calculate average price for each neighborhood
average_prices = {}
for neighborhood in unique_neighborhoods:
    neighborhood_data = air_bnb_df[air_bnb_df['neighbourhood'] == neighborhood]
    average_price = neighborhood_data['price'].mean()
    average_prices[neighborhood] = average_price

# Step 3: Create a DataFrame from the dictionary of average prices
average_prices_df = pd.DataFrame(list(average_prices.items()), columns=['Neighborhood', 'Average Price'])

# Step 4: Display the DataFrame
print(average_prices_df)
````

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap

# Correlation or Relationship analysis between different variables in airbnb dataset

# Exclude non-numeric columns from the correlation computation
numeric_columns = air_bnb_df.select_dtypes(include=['int64', 'float64']).columns
air_bnb_df_numeric = air_bnb_df[numeric_columns]

# Compute the correlation matrix
corr = air_bnb_df_numeric.corr()

# Create the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Columns in Airbnb Data')
plt.show()
````

##### 1. Why did you pick the specific chart?

Visualizing Relationships: Heatmaps are effective for visualizing relationships between variables in a dataset. They provide a visual representation of the correlation between different variables, helping to identify patterns, trends, and dependencies.

Color Encoding: Heatmaps use color encoding to represent the magnitude of values in a matrix. This makes it easy to interpret the data and identify areas of interest. Intensity of color indicates the strength of the relationship between variables.


##### 2. What is/are the insight(s) found from the chart?

In a correlation plot/heat map the values typically range from -1 to 1, representing the strength and direction of the relationship between two variables.A correlation of 1 indicates a perfect positive correlation.This means that as one variable increases, the other variable also increases in a perfectly linear manner.Here we can see that each variable has a diagonal correlation with itself.Values that are close to 1 (e.g., 0.9, 0.95, 0.99) indicate a strong positive correlation. This means that there is a very high likelihood that as one variable increases, the other variable also increases, and the data points are very close to forming a straight line with a positive slope, though not perfectly.Conversely, values close to -1 (e.g., -0.9, -0.95, -0.99) indicate a strong negative correlation, where one variable increases while the other decreases in a near-linear fashion.

Values around 0 indicate little to no linear correlation between the variables, suggesting that there is no clear linear relationship.
